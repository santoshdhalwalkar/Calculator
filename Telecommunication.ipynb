###Unix epoch, using pandas' pd.to_datetime

#### The function has a clear purpose: one for conversion and one for applying that conversion to a DataFrame column.

```python
# Function to convert date string to milliseconds since the Unix epoch using pandas
def to_milliseconds_using_pandas(date_str):
    # Convert date string to datetime, coerce errors to NaT
    timestamp = pd.to_datetime(date_str, format="%m/%d/%Y %H:%M", errors='coerce')
    # Return None if the date is NaT
    if pd.isnull(timestamp):
        return None
    # Convert to milliseconds since Unix epoch
    return int(timestamp.timestamp() * 1000)

# Function to apply the conversion function to a specific column
def convert_column_to_milliseconds(df, column_name):
    df[column_name] = df[column_name].apply(to_milliseconds_using_pandas)

# Convert 'Start' and 'End' columns
convert_column_to_milliseconds(df, 'Start')
convert_column_to_milliseconds(df, 'End')

# Display the first few rows and the updated data types

print(df.dtypes)
```

#### Fill the null values for numerical columns




```python

# Select numerical columnsnumerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Define a function to fill missing values with mean or median
def fill_missing_values(data, columns, strategy='mean'):
    if strategy == 'mean':
        for col in columns:
            df[col].fillna(data[col].mean(), inplace=True)
    elif strategy == 'median':
        for col in columns:
            df[col].fillna(data[col].median(), inplace=True)

            # Example: Filling missing values with median
fill_missing_values(df, numerical_cols, strategy='median')

# Verify that there are no more missing values in the numerical columns
missing_values = df[numerical_cols].isnull().sum()
print(missing_values)
```

Session Duration (Dur. (ms)):

Mean: Average session duration helps in understanding typical usage patterns.
Median: Indicates the midpoint of session durations, offering insight into common session lengths.
Standard Deviation: Helps assess how varied session durations are, which can impact network load.
Range: Shows the shortest and longest sessions, important for network capacity planning.

Data Volume (Total DL (Bytes) and Total UL (Bytes)):

Mean: Average data usage per session, crucial for data plan offerings and capacity planning.
Median: Provides the typical data usage, reducing the skew effect of very high usage sessions.
Standard Deviation: Indicates variability in data usage, impacting how resources are allocated.
Range: Highlights the minimum and maximum data usage, guiding network infrastructure needs.

Round-Trip Time (Avg RTT DL (ms) and Avg RTT UL (ms)):

Mean: Average round-trip time for downloads and uploads, important for assessing network performance.
Median: Gives a typical round-trip time, helping in setting performance benchmarks.
Standard Deviation: Shows the consistency of network performance.
Range: Indicates the best and worst performance, guiding troubleshooting efforts.

Throughput (Avg Bearer TP DL (kbps) and Avg Bearer TP UL (kbps)):

Mean: Average throughput helps in understanding network speed and efficiency.
Median: Typical throughput value, useful for customer experience benchmarks.
Standard Deviation: Shows variability in network speed, impacting quality of service.
Range: Highlights extremes in network performance, crucial for identifying peak load handling.

####Use IQR technique to find and rempove outliers

```python


def iqr_technique(DFcol):
    Q1 = np.percentile(DFcol, 25)
    Q3 = np.percentile(DFcol, 75)

    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
lower_range, upper_range = iqr_technique(df['Durms'])
df = df [(df.Durms > lower_range) & (df.Durms < upper_range) ]

df.shape
```
